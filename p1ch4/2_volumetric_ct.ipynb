`vol_arr` is a 3D tentor.

```python
import imageio

dir_path = "../data/p1ch4/volumetric-dicom/2-LUNG 3.0  B70f-04083"
vol_arr = imageio.volread(dir_path, 'DICOM')
vol_arr.shape
```

The layout is different from what PyTorch expects, due to having no channel information. So we’ll have to make room for the channel dimension using `vol = torch.unsqueeze(vol, 0)`.\
`torch.unsqueeze(input, dim)` takes a tensor (input) and inserts a new dimension of size 1 at the specified position (dim).
`vol`: This is the input tensor. Its shape could be anything, like a 1D vector, 2D matrix, or higher-dimensional tensor.
`0`: This is the dimension index where the new dimension will be added. In PyTorch, dimension indices start at 0, so 0 means the new dimension will be inserted at the very beginning of the tensor’s shape.

```python
vol = torch.from_numpy(vol_arr).float()
vol = torch.unsqueeze(vol, 0)

vol.shape
```
